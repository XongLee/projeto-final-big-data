```python
#pip install beautifulsoup4
from bs4 import BeautifulSoup
from pandas import DataFrame
```

```python
def recuperar_tuplas_de_table(soup, caminho_para_trs_da_table):

    tuplas = []

    headers = []
    
    for indice,linha_da_tabela in enumerate(soup.select(caminho_para_trs_da_table)):
        
        if indice == 0:
            header_columns = linha_da_tabela.findAll('th')
            for header_column in header_columns:
                headers.append(header_column.get_text().strip())
            
        colunas = linha_da_tabela.findAll('td')

        if len(colunas) > 0:

            tupla = dict()
            
            for header, coluna in zip(headers, colunas):
            
                tupla[header] = coluna.get_text().strip()

            tuplas.append(tupla)

            
    return tuplas
```

```python
def recuperar_data_frame_de_tuplas(soup, caminho_para_trs_da_table):
    return DataFrame.from_dict(recuperar_tuplas_de_table(soup, caminho_para_trs_da_table))
```

```python
def recuperar_tuplas_somente_de_aeroportos(soup, caminho_para_trs_da_table):
    tuplas = recuperar_data_frame_de_tuplas(soup, caminho_para_trs_da_table)
    tuplas = tuplas.loc[tuplas['ICAO:'] != '-']
    return tuplas
```

```python
html_file = open('html_geolocation_aeroportos.html', mode='r', encoding='utf-8')
```

```python
soup = BeautifulSoup(html_file, 'html.parser')
```
